# Exploratory Data Analysis (EDA) and Feature Analysis for ECG Classification

This notebook explores the ECG dataset through EDA and feature analysis, examining class distributions, statistical properties, and visualization techniques to gain insights into the data structure before applying classification models.

```python
# ðŸ”¹ Import necessary libraries for data manipulation, visualization, and model training
import pandas as pd  # For handling datasets
import numpy as np  # For numerical operations
from matplotlib import pyplot as plt  # For data visualization
from sklearn.model_selection import train_test_split  # For splitting data into training and testing sets
```

## Data Load and Processing:

```python
# ðŸ”¹ Load each dataset from the specified path
df_train = pd.read_csv("../Data/mitbih_train.csv", header=None)  # Load the training dataset (MIT-BIH)
df_test = pd.read_csv("../Data/mitbih_test.csv", header=None)  # Load the test dataset (MIT-BIH)
```

```python
# ðŸ”¹ Check the dimensions of each dataset
print(f"Training dataset shape: {df_train.shape} (rows, columns)")
print(f"Test dataset shape: {df_test.shape} (rows, columns)")
```

```python
# ðŸ”¹ Check the first few rows of each dataset
print(f"First 5 rows of the Training Dataset:\n{df_train.head()}")  
print(f"\nFirst 5 rows of the Test Dataset:\n{df_test.head()}")  
```

```python
# Check the distribution of labels in each dataset
print(f"Label distribution in Training Dataset:\n{df_train.iloc[:, -1].value_counts()}")  
print(f"\nLabel distribution in Test Dataset:\n{df_test.iloc[:, -1].value_counts()}")  
```

```python
# ðŸ”¹ Compute statistical measures for the training dataset (excluding the label column)
overall_stats = df_train.iloc[:, :-1].describe()

print(f"Statistical Measures for the Training Dataset:\n{overall_stats}")

# ðŸ”¹ Check for missing values in the training dataset
missing_values = df_train.isnull().sum()
print(f"\nMissing values in each column:\n{missing_values[missing_values > 0]}")
```

## Class Distribution:

```python
# ðŸ”¹ Visualize the class distribution in the training dataset with a bar plot
import matplotlib.pyplot as plt

label_distribution = df_train.iloc[:, -1].value_counts()  # Count occurrences of each label in the training set

# ðŸ”¹ Plot the distribution of class labels in the training dataset
plt.figure(figsize=(8,3))
plt.bar(label_distribution.index.astype(str), label_distribution.values, color='skyblue')
plt.xlabel("Class Label")
plt.ylabel("Number of Samples")
plt.title("ðŸ“Š Distribution of Class Labels in the Train Set")
plt.grid(True)

# ðŸ”¹ Save the plot as an image file
plt.savefig("Distribution_of_Class_Labels.png", bbox_inches='tight', dpi=300)

print("Saving and displaying the bar plot for class distribution in the training set:")
plt.show()
```

## Visualization of Time Series for Each Class: 

```python
# ðŸ”¹ Visualize time series ECG signals for each class
# Plot random samples from each class to observe waveform patterns

for label in df_train.iloc[:, -1].unique():
    plt.figure(figsize=(12, 3))

    # ðŸ”¹ Select 5 random ECG signals from the current class
    sample_signals = df_train[df_train.iloc[:, -1] == label].iloc[:, :-1].sample(5, random_state=42)
    
    # ðŸ”¹ Plot each selected ECG signal
    for i, signal in enumerate(sample_signals.values):
        plt.plot(signal, label=f"Sample {i+1}")

    plt.xlabel('Time Step (Sample Index)')
    plt.ylabel('ECG Signal Amplitude')
    plt.title(f'ECG Signal Waveforms for Class {label}')
    plt.legend()
    plt.grid(True)

    # ðŸ”¹ Save the plot as an image file for reference
    plt.savefig(f'ECG_Signal_Waveforms_Class_{label}.png', bbox_inches='tight', dpi=300)

    print(f"Displaying ECG signal waveforms for class {label}:")
    plt.show()
```

## Featrure Analysis: 

```python
# Box plots for feature-wise distribution across all samples
plt.figure(figsize=(15, 8))
plt.boxplot(df_train.iloc[:, :-1].values.T, widths=0.6, showfliers=False)
plt.title('Distribution of ECG Signal Values Across 188 Features')
plt.xlabel('Feature Index (Time Step)')
plt.ylabel('ECG Signal Value')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

## Time Series Characteristics:

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf

# Perform Autocorrelation Analysis for Each Class (0.0 to 4.0)
for label in sorted(df_train.iloc[:, -1].unique()):  # Iterate through each class
    plt.figure(figsize=(10, 3))

    # Select a random sample from the current class
    sample_signal = df_train[df_train.iloc[:, -1] == label].iloc[:, :-1].sample(1, random_state=42).values.flatten()
    
    #  Compute autocorrelation
    autocorr_values = acf(sample_signal, nlags=50, fft=True)

    #  Plot autocorrelation (Fixed: Removed 'use_line_collection' argument)
    plt.stem(range(len(autocorr_values)), autocorr_values, markerfmt='bo', linefmt='b-', basefmt='r-')
    
    plt.xlabel('Lag')
    plt.ylabel('Autocorrelation')
    plt.title(f' Autocorrelation Analysis for Class {label}')
    plt.grid(True)

    # Save the plot as an image file for reference
    plt.savefig(f'Autocorrelation_Class_{label}.png', bbox_inches='tight', dpi=300)

    print(f" Displaying Autocorrelation Plot for Class {label}:")
    plt.show()
```

## Frequency Domain Analysis: 

```python
import numpy as np
import matplotlib.pyplot as plt

# Perform Frequency Domain Analysis for Each Class (0.0 to 4.0)
for label in sorted(df_train.iloc[:, -1].unique()):  # Iterate through each class
    plt.figure(figsize=(8, 4))

    # Select a random ECG signal from the current class
    sample_signal = df_train[df_train.iloc[:, -1] == label].iloc[:, :-1].sample(1, random_state=42).values.flatten()

    # Perform FFT (Fast Fourier Transform)
    fft_values = np.fft.fft(sample_signal)
    fft_freq = np.fft.fftfreq(len(sample_signal), d=1/125)  # Assuming a 125 Hz sampling rate

    # Plot the FFT results (only positive frequencies)
    plt.plot(fft_freq[:len(fft_freq)//2], np.abs(fft_values)[:len(fft_values)//2])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.title(f'Frequency Domain Analysis of Class {label}')
    plt.grid(True)

    # Save the plot as an image file for reference
    plt.savefig(f'Frequency_Domain_Class_{label}.png', bbox_inches='tight', dpi=300)

    print(f"Displaying Frequency Domain Analysis for Class {label}:")
    plt.show()
```

## Feature Engineering for Time Series Data: 

```python
# Add statistical features for each sample
df_features = df_train.copy()  # Create a copy of the original dataset to store new features

# Compute statistical features for each ECG signal
df_features['mean'] = df_train.iloc[:, :-1].mean(axis=1)  # Mean value of the signal
df_features['std'] = df_train.iloc[:, :-1].std(axis=1)  # Standard deviation (measures variability)
df_features['min'] = df_train.iloc[:, :-1].min(axis=1)  # Minimum amplitude value
df_features['max'] = df_train.iloc[:, :-1].max(axis=1)  # Maximum amplitude value
df_features['median'] = df_train.iloc[:, :-1].median(axis=1)  # Median value (robust measure of central tendency)
df_features['iqr'] = df_train.iloc[:, :-1].quantile(0.75, axis=1) - df_train.iloc[:, :-1].quantile(0.25, axis=1)  # Interquartile range (measures spread)
```

## Time-domain Features:

```python
from scipy.stats import skew, kurtosis

# Compute additional time-domain features for each ECG signal
df_features['zero_crossings'] = df_train.iloc[:, :-1].apply(
    lambda x: np.sum(np.diff(np.sign(x)) != 0), axis=1)  # Count the number of times the signal crosses zero

df_features['skewness'] = df_train.iloc[:, :-1].apply(
    lambda x: skew(x), axis=1)  # Measure the asymmetry of the signal distribution

df_features['kurtosis'] = df_train.iloc[:, :-1].apply(
    lambda x: kurtosis(x), axis=1)  # Measure the sharpness/tailedness of the signal distribution
```

## Frequency-Domain Features:

```python
# Add frequency domain features
def fft_features(row):
    """
    Compute FFT-based features for a given signal.
    
    - fft_mean: Mean magnitude of the FFT coefficients.
    - fft_std: Standard deviation of the FFT coefficients (measures variation).
    - fft_max: Maximum magnitude of the FFT coefficients.
    """
    fft_vals = np.fft.fft(row)  # Perform Fast Fourier Transform (FFT)
    fft_mag = np.abs(fft_vals)  # Compute magnitude of FFT coefficients
    return np.mean(fft_mag), np.std(fft_mag), np.max(fft_mag)  # Return computed features

# Apply FFT feature extraction to each ECG signal
df_features[['fft_mean', 'fft_std', 'fft_max']] = df_train.iloc[:, :-1].apply(
    lambda row: fft_features(row), axis=1, result_type='expand'
)
```
